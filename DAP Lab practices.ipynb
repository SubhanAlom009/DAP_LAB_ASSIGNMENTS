```python
# 1. Print "Hello, World!"
print("Hello, World!")

# 2. Calculate the sum of two numbers
num1 = 5
num2 = 7
sum_of_numbers = num1 + num2
print(f"The sum of {num1} and {num2} is: {sum_of_numbers}")

# 3. Find the square of a number
num = 4
square = num ** 2
print(f"The square of {num} is: {square}")

# 4. Accept the user's name and greet them
name = input("Enter your name: ")
print(f"Hello, {name}!")

# 5. Check whether a number is even or odd
num = 3
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")

# 6. Write a Python program that takes a list and returns a new list with unique elements of the first list
input_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = list(set(input_list))
print(f"The unique elements are: {unique_list}")

# 7. Convert Celsius to Fahrenheit
celsius = 25
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}째C is equal to {fahrenheit}째F.")

# 8. Calculate the area of a circle
radius = 5
area_of_circle = 3.14159 * (radius ** 2)
print(f"The area of the circle with radius {radius} is: {area_of_circle}")

# 9. Reverse a given string
input_string = "Python"
reversed_string = input_string[::-1]
print(f"The reversed string is: {reversed_string}")

# 10. Check if a number is a prime number
num = 11
is_prime = True
for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
print(f"{num} is prime: {is_prime}")

# 11. Calculate the factorial of a number
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"The factorial of {num} is: {factorial}")

# 12. Find the largest item from a given list
numbers = [3, 1, 4, 1, 5, 9, 2]
largest_number = max(numbers)
print(f"The largest number in the list is: {largest_number}")

# 13. Write a Python program to check whether a number is in a given range
num = 5
start = 1
end = 10
if start <= num <= end:
    print(f"{num} is within the range of {start} and {end}.")
else:
    print(f"{num} is outside the range of {start} and {end}.")

# 14. Calculate the number of upper case letters and lower case letters in a string
input_string = "Hello World"
upper_case_count = sum(1 for c in input_string if c.isupper())
lower_case_count = sum(1 for c in input_string if c.islower())
print(f"Uppercase letters: {upper_case_count}, Lowercase letters: {lower_case_count}")
```
